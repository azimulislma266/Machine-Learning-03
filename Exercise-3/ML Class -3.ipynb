# Ordinal Encoder

```python
import pandas as pd

df = pd.read_csv('insurance.csv')
df
```

```python
df.smoker.unique()
```

```python
instances = ['yes', 'no']
```

```python
from sklearn.preprocessing import OrdinalEncoder

```

```python
oren = OrdinalEncoder(categories=[instances])
```

```python
df.smoker = oren.fit_transform(df[['smoker']])

df.head()
```
